```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random points for visualization
num_points = 1000
x = np.random.uniform(-1, 1, num_points)
y = np.random.uniform(-1, 1, num_points)

# Check if points are within the circle
radius = 1.0
inside_circle = x**2 + y**2 <= radius**2

# Plot the points
plt.figure(figsize=(6, 6))
plt.scatter(x[inside_circle], y[inside_circle], color='blue', s=5, label='Inside Circle')
plt.scatter(x[~inside_circle], y[~inside_circle], color='red', s=5, label='Outside Circle')
plt.title('Monte Carlo Simulation of Particle Paths')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
